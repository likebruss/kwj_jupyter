```python
# 5.9. Reading Binary Data into a Mutable Buffer
import os.path
def read_into_buffer(filename):
    buf = bytearray(os.path.getsize(filename))
    with open(filename, 'rb') as f:
        f.readinto(buf)
    return buf
# Write a sample file
with open('files/sample.bin', 'wb') as f:
    f.write(b'Hello World')
buf = read_into_buffer('files/sample.bin')
print(buf)
buf[0:5] = b'Hallo'
print(buf)
with open('files/newsample.bin', 'wb') as f:
    f.write(buf)
print()

m1 = memoryview(buf)
m2 = m1[-5:]
m2[:] = b'WORLD'
print(buf)
```

```python
# 5.10. Memory Mapping Binary Files
import os
import mmap

def memory_map(filename, access=mmap.ACCESS_WRITE):
    size = os.path.getsize(filename)
    fd = os.open(filename, os.O_RDWR)
    return mmap.mmap(fd, size, access=access)

size = 1000000
with open('files/data', 'wb') as f:
    f.seek(size-1)
    f.write(b'\x00')
m = memory_map('files/data')
print(len(m))
print(m[0:10])
print(m[0])

# Reassign a slice
m[0:11] = b'Hello World'
m.close()
print()

# Verify that changes were made
with open('files/data', 'rb') as f:
    print(f.read(11))
print()

with memory_map('files/data') as m:
    print(len(m))
    print(m[0:10])
m.closed
```

```python
# 5.18.
from socket import socket, AF_INET, SOCK_STREAM

def echo_client(client_sock, addr):
    print("Got connection from", addr)

    # Make text-mode file wrappers for socket reading/writing
    client_in = open(client_sock.fileno(), 'rt', encoding='latin-1', closefd=False)
    client_out = open(client_sock.fileno(), 'wt', encoding='latin-1', closefd=False)

    # Echo lines back to the client using file I/O
    for line in client_in:
        client_out.write(line)
        client_out.flush()
    client_sock.close()

def echo_server(address):
    sock = socket(AF_INET, SOCK_STREAM)
    sock.bind(address)
    sock.listen(1)
    while True:
        client, addr = sock.accept()
        echo_client(client, addr)

if __name__ == '__main__':
    print('Echo serving running on localhost:25000')
    echo_server(('', 25000))
```

```python
# 5.21.1
# class that internally defines a thread but can still be pickled/unpickled
# countdown.py
import time
import threading
class Countdown:
    def __init__(self, n):
        self.n = n
        self.thr = threading.Thread(target=self.run)
        self.thr.daemon = True
        self.thr.start()
    def run(self):
        while self.n > 0:
            print('T-minus', self.n)
            self.n -= 1
            time.sleep(5)
    def __getstate__(self):
        return self.n
    def __setstate__(self, n):
        self.__init__(n)
        
import countdown
c = Countdown(3)
```

```python
# 5.21.2
# After a few moments
f = open('files/cstate.p', 'wb')
import pickle
pickle.dump(c, f)
f.close()
```
